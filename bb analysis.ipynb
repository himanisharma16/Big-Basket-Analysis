Step 7: Find out the Missing Values from the Dataset.



```python
data.isnull().sum()
```

```python
#percentage of missing values 
round((data.isnull().sum()/(len(data)))*100,2)
```

```python
#Data Reduction
data = data.drop(['index'], axis = 1)
data.info()
```

```python
# column rating has high percentage of missing values therefore it is better to drop
data=data.drop(columns=['rating'],errors='ignore')
```

```python
#imputing for missing values
#for numeric columns
data['sale_price']=data['sale_price'].fillna(data['sale_price'].median())
```

```python
#missing value in categorical column
data=data.dropna(subset=['product','brand'])
```

```python
# handling null value in description with "No Description"
data['description']=data['description'].fillna('No Description')
```

```python
data.isnull().sum()
```

Step 8: Find out the outliers from the dataset according to the columns and fill them with the mean.

```python
# column 'sale_price'
Q1 = data['sale_price'].quantile(0.25)
Q3 = data['sale_price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Replace outliers with mean
data['sale_price'] = np.where(
    (data['sale_price'] < lower_bound) | (data['sale_price'] > upper_bound),data['sale_price'].mean(),data['sale_price'])

```

```python
# column 'market_price'
Q1 = data['market_price'].quantile(0.25)
Q3 = data['market_price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound_m = Q1 - 1.5 * IQR
upper_bound_m = Q3 + 1.5 * IQR

# Replace outliers with mean
data['market_price'] = np.where(
    (data['market_price'] < lower_bound) | (data['market_price'] > upper_bound),data['market_price'].mean(),data['market_price'])
```

Step 9: Create Plots or visualizations.

```python

# Scatter plot for Sale Price vs Market Price
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sale_price', y='market_price', data=data)
plt.title('Sale Price vs Market Price')
plt.xlabel('Sale Price')
plt.ylabel('Market Price')
plt.show()
```

```python
# Box plot for numerical column 'Sale Price' vs categorical column 'Category'
plt.figure(figsize=(8, 6))
sns.boxplot(x='category', y='sale_price', data=data)
plt.title('Sale Price Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Sale Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# top 10 most costly brand according to sale price
plt.subplots(figsize=(12,7))
data.groupby('brand')['sale_price'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar') 
```
